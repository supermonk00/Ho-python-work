####轉換資料變成浮點數的方法

```python
import locale 

# atof
```

```python
locale.setlocale(locale.LC_NUMERIC,"")
```

```python
df2["男生計"] = df2['男生計'].apply(locale.atof)
```

```python
df2["女生計"] = df2['女生計'].apply(locale.atof)
```
