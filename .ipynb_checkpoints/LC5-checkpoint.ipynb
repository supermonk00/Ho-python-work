```python
x = np.linspace(0,5,50)
```

```python
X  = x.reshape(50,1)
```

```python
y = 1.2*x + 0.8 + 0.5*np.random.randn(50)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()
```

```python
regr.fit(X,y)
```

```python
Y = regr.predict(X)
```

```python
regr.predict([[1.3]])
```

```python
plt.scatter(x,y)
plt.plot(x,Y, "r")
plt.plot(x, 1.2*x + 0.8 , "b")
```
